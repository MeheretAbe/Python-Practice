### Task 5
Thomas and several other customers seem to be interested in cars with black or gray colored interiors. John wants to figure out if the cars with interiors in black or gray are priced higher or lower than cars having interiors of other colors.
Determine the average price of all the 100 lowest-priced cars from the year 2014 based on their interior colors.

```python
# all cars with interiors in black or gray with lower to higher mps
I=T[(T['interior'] == "black") | (T['interior'] == "gray")&
     (T.state=="ca")].sort_values(by='mp', ascending=True)
I
```

```python
#count of all cars with interiors in black or gray
pd.pivot_table(I, index=['interior'], values=['mp'], aggfunc='count')
```

```python
'''
the average price of all the 100 lowest-priced cars from the year 2014 
based on their interior colors
'''

Y= T[T['year'] == 2014]
pd.pivot_table(Y, index='interior', values='mp', aggfunc='mean').sort_values(by='mp', ascending=False)
```

### Task 7
Eric, another customer, is interested in a good-condition BMW Sedan manufactured between 2010 and 2015. John wants to display the details of all such cars that have a condition score >=4.5.
Help John to figure out the year-wise maximum expected price (mp) of the BMW cars manufactured between the years 2010 and 2015 grouped by each model.

```python
E= df[(df.year>= 2010) &
      (df.year< 2015) &
      (df.make == 'BMW') &
      (df.condition >= 4.5)&
      (df.body == 'Sedan')]

E
```

```python
# year-wise maximum expected price (mp) of the BMW cars
    
    
pd.pivot_table(E, index='model', columns='year', values='mp', aggfunc='mean')
```

```python
#or 
E.groupby(['model','year'])['mp'].mean()
```
